### Storing information about all the students who applied for a course

```python
coursewise_ta_dict = {}

for each in courses_dict:
    coursewise_ta_dict[each] = []

for course in courses_dict:
    for ta in range(total_TAs):
        TA = df.iloc[ta]
        for num in range(1, 5+1):
            offer = courses_dict[course]['offered_for']
            TA_rollno = TA['Roll No.']
            if not pd.isna(TA[f'Preference-{num}']) and TA[f'Preference-{num}'] == course and check_offered_for(TA_rollno, offer):
                if num == 1:
                    # if check_grade(TA['Grade']):
                        TA_program_prefence = get_program_preference(TA_rollno)
                        TA_course_preference = (6-num)
                        TA_info = [TA_program_prefence, TA_course_preference, TA_rollno, TA['Grade']]
                        coursewise_ta_dict[course].append(TA_info)
                else:
                    # if check_grade(TA[f'Grade.{num-1}']):
                        TA_program_prefence = get_program_preference(TA_rollno)
                        TA_course_preference = (6-num)
                        TA_info = [TA_program_prefence, TA_course_preference, TA_rollno, TA[f'Grade.{num-1}']]
                        coursewise_ta_dict[course].append(TA_info)

coursewise_ta_dict
```

### Generating possible combinations of TAs based on the constraints for each course

```python
course_TA_combos = {}

for course in courses_dict:
    required = courses_dict[course]['num_TAs']
    num_students = courses_dict[course]['num_students']
    all_TAs = sorted(coursewise_ta_dict[course], reverse=True)
    combo = generate_combinations(all_TAs, required)
        
    combo[:] = [each for each in combo if (check_ug(each) and check_pg(each, num_students) and (num_students > 20) and check_grade(each))]
    
    course_TA_combos[course] = combo
```

## Constraint Satisfaction Problem
---

```python
def assign(rollno, course_name, TAs_assigned):
    TAs_assigned[rollno] = course_name
    return TAs_assigned
    
def unassign(rollno, TAs_assigned):
    del TAs_assigned[rollno]
    return TAs_assigned
```

```python
num_courses = len(courses)
num_courses
```

```python
def check_all_assigned(combo, TAs_assigned):
    for each in combo:
        if each in TAs_assigned:
            return False
    return True
```

```python
def CSP(index, TAs_assigned):
    if index == num_courses:
        return TAs_assigned
    
    course_name = course_num_dict[index]
    
    if len(course_TA_combos[course_name]) == 0:
        res = CSP(index+1, TAs_assigned)
        if res is not None:
            return res
    
    for each in range(len(course_TA_combos[course_name])):
        possible_TAs = np.array(course_TA_combos[course_name][each])
        if len(possible_TAs) == 0:
            rollnos = []
        else:
            rollnos = possible_TAs.T[2]
        
        if check_all_assigned(rollnos, TAs_assigned):
            for ta in rollnos:
                TAs_assigned = assign(ta, course_name, TAs_assigned)
            
            res = CSP(index+1, TAs_assigned)
            if res is not None:
                return res
            
            for ta in rollnos:
                TAs_assigned = unassign(ta, TAs_assigned)
                
    return None
```

## Assigned TAs

```python
assigned_TAs = []

for rollno in res:
    for each in range(len(df)):
        ta = df.iloc[each]
        ta_rollno = ta['Roll No.']
        if rollno == ta_rollno:
            name = ta['Name']
            program = ta['Program']
            branch = ta['Branch']
            course = res[rollno]
            
            for num in range(1, 5+1):
                if ta[f'Preference-{num}'] == course:
                    course_preference = num
                    if num == 1:
                        info = [course, rollno, name, program, branch, course_preference, ta['Grade']]
                        assigned_TAs.append(info)
                        break
                    else:
                        info = [course, rollno, name, program, branch, course_preference, ta[f'Grade.{num-1}']]
                        assigned_TAs.append(info)                    
                        break
            break

# assigned_TAs
```

```python
assigned_TA_df = pd.DataFrame(assigned_TAs, columns=['Course', 'Roll No.', 'Name', 'Program', 'Branch', 'Preference Number', 'Grade'])
assigned_TA_df
```

```python
total_assigned_TAs = len(assigned_TA_df)
for each in range(1, 5+1):
    if each == 1:
        word = '1st'
    elif each == 2:
        word = '2nd'
    elif each == '3':
        word = '3rd'
    else:
        word = f'{each}th'
    count = len(assigned_TA_df[assigned_TA_df['Preference Number']==each])
    percentage = count / len(assigned_TA_df)
    print(f'Number of assigned TAs who got their {word} preference: {count} ({round(percentage*100)}%)')
```
